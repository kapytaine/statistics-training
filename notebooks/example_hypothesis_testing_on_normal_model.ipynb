# Packages

```python
%load_ext nb_black

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2, t, multivariate_normal
```

___
# Data
Let consider the following example with the following student heigh measures.

```python
x = np.array([1.55, 1.75, 1.89, 1.96])
```

```python
n = len(x)
```

___
# Model

Let assume:
- $(X_i)_{i \in \mathbb{N}}$ i.i.d.
- $X_1 \sim \mathcal{N}(\mu, \sigma^2)$

___
# Likelihood
The likelihood is a function of the normal distribution parameters, $\mu$ and $\sigma$, considering the realizations of the random variables as fixed.
$$L(\mu, \sigma ; x_1,...,x_n) = f_{(X_1, ..., X_n)}(x_1, ..., x_n)$$

Since we assume that $(X_i)_{i \in \{1..n\}}$ are i.i.d. random variables, then:
$$f_{(X_1, ..., X_n)}(x_1, ..., x_n) = f_{X_1}(x_1)...f_{X_n}(x_n)$$

The log likelihood is then:
$$log(L(\mu, \sigma ; x_1,...,x_n)) = \sum_{i=1}^{n}log(f_{X_i}(x_i))$$

```python
fig, ax = plt.subplots(figsize=(6, 6))

mu = np.linspace(1.5, 2.1, 100)
sigma = np.linspace(0.01, 0.6, 100)
X, Y = np.meshgrid(mu, sigma)
Z = np.zeros_like(X)
for ii in range(Z.shape[0]):
    for jj in range(Z.shape[1]):
        rv = multivariate_normal([mu[jj]] * n, sigma[ii] ** 2 * np.eye(n))
        Z[ii, jj] = rv.pdf(x)
CS = ax.contour(X, Y, Z, levels=10)
ax.clabel(CS, inline=True, fontsize=10)
ax.set_xlabel("")
ax.set_ylabel("")
ax.set_title("Isolevel lines of the likelihood")

mu_hat = np.mean(x)
sigma_hat = np.std(x, ddof=1)
ax.plot(mu_hat, sigma_hat * np.sqrt((n - 1) / n), "r+", label="MLE")
ax.legend(loc="best")
```

# Estimators

## Parameter

```python
alpha = 0.05
```

## Mean

### Moment method
The first moment is enough to construct an estimator of $\mu$.
$$\mathbb{E}[X_1]=\mu \iff \mu=\mathbb{E}[X_1]$$
We can then construct the following estimator $\hat{\mu}_{MM} = \bar{X_n}$. It is:

#### Bias
The estimator is **unbiased**.

#### Consistency
Let's apply the weak law of large number (LLN).
$$\left( \begin{cases}
\mathbb{E}[X_1^2]=\sigma^2+\mu^2<+\infty \\
(X_i)_{i\in{1..n}} \quad \text{i.i.d. random variables} \end{cases} \right) \Rightarrow \bar{X_n} \overset{\mathbb{P}}{\to}\mathbb{E}[X_1]=\mu$$
The estimator is **consistent**.

#### Convergence
There is no need to apply the CLT because we can directly calculate the distribution of $\sqrt{n}(\bar{X_n} - \mu) \sim \mathcal{N}(0, \sigma^2)$.
The estimator is **convergent** with rate $\sqrt{n}$.

### MLE
$$\begin{aligned}
\hat{\begin{bmatrix}
\mu \\
\sigma \end{bmatrix}}_{MLE}(x_1, ..., x_n) &= \underset{(\mu, \sigma) \in  \mathbb{R} \times \mathbb{R^{+*}}}{\mathrm{argmax}} L(\mu, \sigma ; x_1, ..., x_n) \\
&= \begin{bmatrix}
\bar{x_n} \\
\frac{1}{n} \sum_{i=1}^{n}(x_i-\bar{x_n})^2 \end{bmatrix}
\end{aligned}$$
Hence, the estimator of the mean is $\hat{\mu}_{MLE} = \bar{X_n}$. It is the same as the estimator obtained by the moment method.

### Confidence interval
The confidence interval of the mean can be computed by means of the following pivot function $g(X, \mu) = \sqrt{n}\frac{\bar{X_n}-\mu}{\sqrt{\frac{1}{n-1} \sum_{i=1}^{n}(X_i-\bar{X_n})^2}}$ which is a statistic that follows a distribution independently of the unknown parameters: $g(X, \mu) \sim \mathcal{T}(n-1)$.

```python
mu = np.linspace(1, 2.5, 1000)
mu_hat = np.mean(x)
sigma_hat = np.std(x, ddof=1)

fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.plot(
    mu,
    t.pdf((mu - mu_hat) / (sigma_hat / np.sqrt(n)), n - 1),
    label="density of g(X,) at (x1, ..., xn)",
)
ax.vlines(
    t.ppf([alpha / 2, 1 - alpha / 2], n - 1) * sigma_hat / np.sqrt(n) + mu_hat,
    ymin=0,
    ymax=0.35,
    label="confidence interval bounds of coverage 1-",
    color="tab:orange",
)
ax.set_xlabel("")
ax.legend(loc="best")
```

## Variance

### Moment method
To construct an estimator of $\sigma^2$ by means of the moment method, we need to compute the first two moments. 
$$\begin{aligned}
\begin{bmatrix} \mathbb{E}[X_1] \\ \mathbb{E}[X_1^2] \end{bmatrix} &= \begin{bmatrix} \mu \\ \sigma^2 + \mu^2 \end{bmatrix} \\
\iff \begin{bmatrix} \mu \\ \sigma^2 \end{bmatrix} &= \begin{bmatrix} \mathbb{E}[X_1] \\ \mathbb{E}[X_1^2] - \mathbb{E}[X_1]^2 \end{bmatrix}
\end{aligned}$$
Consequently the estimator of $\sigma^2$ constructed by the moment method is: $\hat{\sigma}_{MM} = \bar{X_n^2} - \bar{X_n}^2$.

#### Bias
$\mathbb{E}[\hat{\sigma}_{MM}] = \mathbb{E}[X_1^2] - (Var[\bar{X_n}] + \mathbb{E}[\bar{X_n}]^2) = (\sigma^2+\mu^2) - (\frac{\sigma^2}{n} + \mu^2) = \frac{n-1}{n}\sigma^2$, so the estimator is **biased**.

#### Consistency
Let's first proove that $\bar{X_n}^2$ converges.
$$\left( \begin{cases}
\bar{X_n} \overset{\mathbb{P}}{\to}\mu \Rightarrow \bar{X_n} \overset{\mathcal{L}}{\to}\mu \\
g(x)=x^2 \quad \text{is a measurable and continuous function} \end{cases} \right) \Rightarrow g(\bar{X_n})=\bar{X_n}^2 \overset{\mathcal{L}}{\to} g(\mathbb{E}[X_1])=\mathbb{E}[X_1]^2=\mu^2$$

Let's then proove the convergence of the second term. We can apply the strong LLN.
$$\left( \begin{cases}
\mathbb{E}[|X_1^2|]=\mathbb{E}[X_1^2]=\sigma^2+\mu^2<+\infty \\
(X_i^2)_{i\in{1..n}} \quad \text{are i.i.d. random variables} \end{cases} \right) \Rightarrow \bar{X_n^2} \overset{\mathcal{a.s.}}{\to} \mathbb{E}[X_1^2]=\sigma^2+\mu^2$$

Thanks to Slutsky, we have that:
$$(\bar{X_n}^2, \bar{X_n^2}) \overset{\mathcal{L}}{\to} (\mu^2, \sigma^2+\mu^2)$$

Finally, we use the convergence by composition:
$$\left( \begin{cases}
(\bar{X_n}^2, \bar{X_n^2}) \overset{\mathcal{L}}{\to} (\mu^2, \sigma^2+\mu^2) \\
g(x,y)=y-x \quad \text{is measurable and continuous over } \mathbb{R} \times \mathbb{R} \end{cases} \right) \Rightarrow g(\bar{X_n}^2, \bar{X_n^2}) \overset{\mathcal{L}}{\to} g(\mu^2, \sigma^2+\mu^2) =  \sigma^2$$
Since it is a constant, it also converges in probability, hence the **consistency**.

#### Convergence
Thanks to Cochran theorem, we know that $\frac{n}{\sigma^2}(\bar{X_n^2}-\bar{X_n}^2) = \frac{1}{\sigma^2}\sum_{i=1}^{n}(X_i-\bar{X_n})^2 \sim \chi^2(n-1) = \Gamma(\frac{1}{2}, \frac{n-1}{2})$. Scaling with $n-1$, we have: $\frac{n}{(n-1)\sigma^2}(\bar{X_n^2}-\bar{X_n}^2) \sim \Gamma(\frac{n-1}{2}, \frac{n-1}{2})$. The characteristic function of this later distribution is known as well as the function it converges to. We finally have the following equivalence:
$$\left( \phi_n(x) = \left( \frac{\frac{n-1}{2}}{\frac{n-1}{2}-iu} \right)^{\frac{n-1}{2}} \to e^{iu} \quad \forall u \in \mathbb{R} \right) \iff \left( \frac{n}{(n-1)\sigma^2}(\bar{X_n^2}-\bar{X_n}^2) \overset{\mathcal{L}}{\to} X \quad L(X)=\delta_1 \right)$$


### MLE
Thanks to the previous calculation, we already know the MLE of $\sigma^2$, $\hat{\sigma^2}_{MLE} = \frac{1}{n} \sum_{k=1}^{n} (X_i-\bar{X_n})^2 = \bar{X_n^2} - \bar{X_n}^2 = \hat{\sigma}_{MM}$.

#### Bias
Biased, see previously.

#### Consistency
Consistent, see previously.

#### Convergence
We could have a better rate thanks to the convergence of the maximum likelihood estimator.
$$
\left( \text{the model is regular} \right)
\Rightarrow
\left( \sqrt{n}(\hat{\sigma^2}_{MLE} - \sigma^2) \overset{\mathcal{L}}{\to} \mathcal{N}(0, I(\sigma)^{-1}) \right)
$$
We can calculate the Fisher information: $I(\sigma^2) = \frac{1}{2\sigma^4}$. Finally, we have:
$$\begin{aligned}
\sqrt{n}(\hat{\sigma^2}_{MLE} - \sigma^2) & \overset{\mathcal{L}}{\to} \mathcal{N}(0, 2\sigma^4) \\
\iff \frac{\sqrt{n}(\hat{\sigma^2}_{MLE} - \sigma^2)}{\sqrt{2}\sigma^2} & \overset{\mathcal{L}}{\to} \mathcal{N}(0, 1)
\end{aligned}$$

We then have a better convergence rate $\sqrt{n}$ instead of $1$ obtained in previous chapter.

#### Confidence interval
The confidence interval can be computed by means of the following pivot function $g(X, \sigma^2) = \frac{1}{\sigma^2}\sum_{i=1}^{n}(X_i-\bar{X_n})^2$ which is a statistic that follows a distribution independently of the unknown parameters: $g(X, \sigma^2) \sim \chi^2(n-1)$.

```python
sigma = np.linspace(0.001, 1, 1000, endpoint=False)
mu_hat = np.mean(x)
sigma_hat = np.std(x, ddof=1)

fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.plot(
    sigma,
    chi2.pdf((n - 1) * sigma_hat**2 / (sigma**2), n - 1),
    label="density of g(X,) at (x1, ..., xn)",
)
ax.vlines(
    np.sqrt((n - 1) / chi2.ppf([alpha / 2, 1 - alpha / 2], n - 1) * sigma_hat**2),
    ymin=0,
    ymax=0.35,
    label="confidence interval bounds of coverage 1-",
    color="tab:orange",
)
ax.set_xlabel("")
ax.legend(loc="best")
```

# Hypothesis testing

## Example 1

H0: $\mu \ge \mu_0 = 2$

H1: $\mu < \mu_0 = 2$

```python
mu_0 = 2
```

```python
mu_hat = np.mean(x)
sigma_hat = np.std(x, ddof=1)
T = np.sqrt(n) * (mu_hat - mu_0) / sigma_hat
```

```python
if T < t.ppf(alpha, n - 1):
    print("Reject H0")
else:
    print("Accept H0")
```

```python
mu = np.linspace(0, 2)
error_2nd = t.sf(t.ppf(alpha, n - 1) + np.sqrt(n) / sigma_hat * (mu_0 - mu), n - 1)

fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.plot(mu, error_2nd, label="type II error")
ax.set_xlabel("")
ax.legend(loc="best")
```
